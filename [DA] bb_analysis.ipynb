```python
from pathlib import Path

apply_fn= lambda x: str(
                        Path(x.replace("{XAMI_MIMIC_PATH}", XAMI_MIMIC_PATH))
                    )
```

```python
d = "Pleural abnormality"
```

```python
reflacx_df['anomaly_location_ellipses_path'] = reflacx_df['anomaly_location_ellipses_path'].apply(apply_fn)
```

```python
all_ab_bb = []
for _, i in reflacx_df[reflacx_df[d] > 0].iterrows():
    e_df = pd.read_csv(i['anomaly_location_ellipses_path'])
    if d in e_df.keys():
        e_df["xmin"] = e_df["xmin"] /  i['image_size_x']
        e_df["ymin"] = e_df["ymin"] /  i['image_size_y']
        e_df["xmax"] = e_df["xmax"] /  i['image_size_x']
        e_df["ymax"] = e_df["ymax"] /  i['image_size_y']
        all_ab_bb.append(e_df[e_df[d]])
```

```python
ab_bb_df=pd.concat(all_ab_bb, axis=0)
```
